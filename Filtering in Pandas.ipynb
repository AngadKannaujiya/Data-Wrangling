# Grouping the data #  

By using double square brackets, we are selecting 'Kms_Driven' as a DataFrame rather than a Series, which can be useful for consistency when selecting multiple columns.

```python
cardata.groupby('Fuel_Type')[['Kms_Driven', 'Present_Price']].mean() 
```

```python
cardata.groupby('Fuel_Type')['Kms_Driven'].agg(['mean', 'median',])
```

```python
cardata.groupby('Fuel_Type').count()
```
